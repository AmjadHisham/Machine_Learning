```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

pd.set_option('max_columns', 50)
mpl.rcParams['lines.linewidth'] = 2
%matplotlib inline
```

```python
df = pd.read_csv('relay-foods.csv')
df.dtypes
```

```python
df['OrderDate'] = df.OrderDate.apply(lambda x: pd.datetime.strptime(x, '%m/%d/%Y'))
df['PickupDate'] = df.PickupDate.apply(lambda x: pd.datetime.strptime(x, '%m/%d/%Y'))
df.head()
```

```python
df['OrderPeriod'] = df.OrderDate.apply(lambda x: x.strftime('%Y-%m'))
df.head()
```

```python
df.set_index('UserId', inplace=True)

df['CohortGroup'] = df.groupby(level=0)['OrderDate'].min().apply(lambda x: pd.datetime.strftime(x, '%Y-%m'))
df.reset_index(inplace=True)
df.head()
```

```python
grouped = df.groupby(['CohortGroup','OrderPeriod'])

cohorts = grouped.agg({
        'UserId' : pd.Series.nunique,
        'OrderId' : pd.Series.nunique,
        'TotalCharges' : np.sum
    })

cohorts.rename(columns={
        'UserId' : 'TotalUsers',
        'OrderId' : 'TotalOrders'
    }, inplace=True)
cohorts.head(16)
```

```python
def cohort_period(df):
    df['CohortPeriod'] = np.arange(len(df)) + 1
    return df

cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts.head(20)
```

```python
x = df[(df.CohortGroup == '2009-01') & (df.OrderPeriod == '2009-01')]
y = cohorts.ix[('2009-01', '2009-01')]

assert(x['UserId']).nunique() == y['TotalUsers']
assert(x['TotalCharges'].sum().round(2) == y['TotalCharges'].round(2))
assert(x['OrderId'].nunique() == y['TotalOrders'])

x = df[(df.CohortGroup == '2009-01') & (df.OrderPeriod == '2009-09')]
y = cohorts.ix[('2009-01', '2009-09')]

assert(x['UserId'].nunique() == y['TotalUsers'])
assert(x['TotalCharges'].sum().round(2) == y['TotalCharges'].round(2))
assert(x['OrderId'].nunique() == y['TotalOrders'])

x = df[(df.CohortGroup == '2009-05') & (df.OrderPeriod == '2009-09')]
y = cohorts.ix[('2009-05', '2009-09')]

assert(x['UserId'].nunique() == y['TotalUsers'])
assert(x['TotalCharges'].sum().round(2) == y['TotalCharges'].round(2))
assert(x['OrderId'].nunique() == y['TotalOrders'])
```

```python
cohorts.reset_index(inplace=True)
cohorts.set_index(['CohortGroup','CohortPeriod'], inplace=True)

cohorts_group_size = cohorts['TotalUsers'].groupby(level=0).first()
cohorts_group_size.head(10)
```

```python
cohorts['TotalUsers'].head(10)
```

```python
cohorts['TotalUsers'].unstack(0).head()
```

```python
user_retention = cohorts['TotalUsers'].unstack(0).divide(cohorts_group_size, axis=1)
user_retention.head(10)
```

```python
user_retention[['2009-06', '2009-07', '2009-08']].plot(figsize=(10,5))
plt.title('Cohorts: User Retention')
plt.xticks(np.arange(1, 12.1, 1))
plt.xlim(1, 12)
plt.ylabel('% of Cohort Purchasing');
```

```python
import seaborn as sns
sns.set_style('white')
```
