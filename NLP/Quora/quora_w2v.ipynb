```python
import sys
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
quora = pd.read_csv('quora_duplicate_questions.tsv', delimiter='\t')
```

```python
quora.to_csv('q_quora.csv')
```

```python
quora.head(10)
```

```python
quora['question1'] = quora['question1'].apply(lambda x: unicode(str(x), 'utf-8'))
quora['question2'] = quora['question2'].apply(lambda x: unicode(str(x), 'utf-8'))
```

```python
quora.dtypes
```

```python
import gensim

questions = list(quora['question1']) + list(quora['question2'])
```

```python
len(questions)
```

```python
#tokenize

c = 0 
for question in tqdm(questions):
    questions[c] = list(gensim.utils.tokenize(question, deacc=True, lower=True))
    c += 1
```

```python
len(questions)
```

```python
questions[0]
```

```python
# train model
model = gensim.models.Word2Vec(questions, size=300, workers=16, iter=10, negative=20)
```

```python
# trim memory
model.init_sims(replace=True)
```

```python
# creta a dict 
w2v = dict(zip(model.index2word, model.syn0))
print "Number of tokens in Word2Vec:", len(w2v.keys())
```

```python
w2v.keys()
```

```python
# save model
model.save('3_word2vec.mdl')
model.save_word2vec_format('3_word2vec.bin', binary=True)
```

```python
model.most_similar('woods')
```

```python
from nltk import *

words = np.concatenate(questions[:20000])
```

```python
len(words)
```

```python
fdist = FreqDist(words)
fdist
```

```python
common = fdist.most_common(1000)
common
```

```python
words_top_quora = []
for count, words in enumerate(common):
    words_top_quora.append(str(words[0]))
```

```python
words_top_quora[900:]
```

```python
words_vec = model[words_top_quora]
words_vec.shape
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)
words_top_quora_tsne = tsne.fit_transform(words_vec)
```

```python
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
output_notebook()
```

```python
p = figure(tools="pan,wheel_zoom,reset,save",
           toolbar_location="above",
           title="word2vec T-SNE for most common words")

source = ColumnDataSource(data=dict(x1=words_top_quora_tsne[:,0],
                                    x2=words_top_quora_tsne[:,1],
                                    names=words_top_quora))

p.scatter(x="x1", y="x2", size=8, source=source)

labels = LabelSet(x="x1", y="x2", text="names", y_offset=6,
                  text_font_size="8pt", text_color="#555555",
                  source=source, text_align='center')
p.add_layout(labels)

show(p)
```
