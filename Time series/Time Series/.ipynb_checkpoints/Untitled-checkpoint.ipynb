```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20, 8
%matplotlib inline
```

```python
data = pd.read_csv('Air.csv')
print data.head()
print '\n data type:'
print data.dtypes
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
ar_data = pd.read_csv('Air.csv',index_col='Month',date_parser=dateparse)
print ar_data.head()
print '\n Data type :'
print ar_data.dtypes
```

```python
ar_data.index
```

```python
plt.plot(ar_data)
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(ar_data.values.ravel())
```

```python
ts_log = np.log(ar_data)
plt.plot(ts_log)
```

```python
movng_avg = pd.rolling_mean(ts_log, 12)
plt.plot(ts_log)
plt.plot(movng_avg, color='red')
```

```python
ts_log_mov_avg_diff = ts_log - movng_avg
ts_log_mov_avg_diff.head(14)
```

```python
ts_log_mov_avg_diff.dropna(inplace=True)
test_stationarity(ts_log_mov_avg_diff.values.ravel())
```

```python
exp_weight = pd.ewma(ts_log, halflife=12)
plt.plot(ts_log)
plt.plot(exp_weight, color='red')
```

```python
ts_exp_diff = ts_log - exp_weight
test_stationarity(ts_exp_diff.values.ravel())
```

```python
ts_sim_diff = ts_log - ts_log.shift()
plt.plot(ts_sim_diff)
```

```python
ts_sim_diff.dropna(inplace=True)
test_stationarity(ts_sim_diff.values.ravel())
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(ts_log)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(ts_log, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose.values.ravel())
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(ts_sim_diff, nlags=20)
lag_pacf = pacf(ts_sim_diff, nlags=20, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_sim_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_sim_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_sim_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_sim_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
#AR Model
model = ARIMA(ts_log, order=(2, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(ts_sim_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_sim_diff.values.ravel())**2))
```

```python
#MA model
model = ARIMA(ts_log, order=(0, 1, 2))  
results_MA = model.fit(disp=-1)  
plt.plot(ts_sim_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_sim_diff.values.ravel())**2))
```

```python
#combined Model
model = ARIMA(ts_log, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.plot(ts_sim_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_sim_diff.values.ravel())**2))
```

```python
pred_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print pred_ARIMA_diff.head()
```

```python
pred_ARIMA_diff_cumsum = pred_ARIMA_diff.cumsum()
print pred_ARIMA_diff_cumsum.head()
```

```python
predictions_ARIMA_log = pd.Series(ts_log.ix[0].ravel(), index=ts_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(pred_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
ts_log.ix[:]
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(ar_data)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-ar_data.values.ravel())**2)/len(ar_data)))
```
