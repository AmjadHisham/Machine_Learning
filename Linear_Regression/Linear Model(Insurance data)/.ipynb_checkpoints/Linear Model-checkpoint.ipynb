## Creating dummy variables
    creating text var to numeric using dummy variable concept
    Gender: M(1)/F(0)
    Married: Married(1)/Single(0)
    FuelType: P(1)/D(0)

```python
# creating dummy variable using map function
data['D_Gender'] = data['Gender'].map({'M':1,'F':0})
data['D_Married'] = data['Married'].map({'Married':1,'Single':0})
data['D_FuelType'] = data['FuelType'].map({'P':1,'D':0})
```

### Feature Selection

```python
from sklearn.feature_selection import RFE, f_regression
from sklearn.linear_model import LinearRegression, RandomizedLasso, LassoCV
```

```python
X = data[['Age','DrvExp','VehNum','VehAge','D_Gender','D_Married','D_FuelType']]
Y = data['Losses']
```

### Ranking variables according to Linear Model

```python
lr = LinearRegression()

rfe = RFE(lr, n_features_to_select=1)
rfe.fit(X,Y)
```

```python
print "Features sorted by their rank:"
print sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), X.columns)) 
```

## Linear Model

```python
from sklearn.cross_validation import train_test_split
```

```python
Xn = data[['Age','DrvExp','VehAge','D_Gender','D_Married','D_FuelType']]
Yn = data['Losses']
```

```python
X_train, x_test, Y_train, y_test = train_test_split(Xn,Yn, test_size=0.33, random_state=42)
X_train.shape, Y_train.shape
```

```python
lr.fit(X_train,Y_train)
```

```python
print "MSE for model :" ,np.sqrt(np.mean((Y_train - lr.predict(X_train))**2))
```

```python
print "MSE for testing sample :" ,np.sqrt(np.mean((y_test - lr.predict(x_test))**2))
```

```python
resd = (y_test - lr.predict(x_test))
sns.residplot(y_test, resd)
```

```python
plt.scatter(y_test, lr.predict(x_test))
```

```python
def pretty_print_linear(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
```

```python
pretty_print_linear(lr.coef_)
```

## GradientBostingRegressor Model

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
params = {'n_estimators': 800, 'max_depth': 6, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}

gbr = GradientBoostingRegressor(**params)

gbr.fit(X_train,Y_train)
```

```python
print "MSE of Model :" ,np.sqrt(np.mean((Y_train - gbr.predict(X_train))**2))
```

```python
print "MSE of test sample :" ,np.sqrt(np.mean((y_test - gbr.predict(x_test))**2))
```

```python
plt.scatter(y_test, gbr.predict(x_test))
```

```python
Gresd = (y_test - gbr.predict(x_test))
sns.residplot(y_test, Gresd)
```

```python
feature_importance = gbr.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, Xn.columns)
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

```python
np.sqrt(np.mean((Yn - gbr.predict(Xn))**2))
```
